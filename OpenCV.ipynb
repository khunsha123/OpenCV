```python
import cv2
from cv2 import dnn_superres

# Create an SR object
sr = dnn_superres.DnnSuperResImpl_create()

# Read image
image = cv2.imread('/content/TF-LapSRN/Test/t13.png')

# Read the desired model
path = "/content/TF-LapSRN/export/LapSRN_x2.pb"
sr.readModel(path)

# Set the desired model and scale to get correct pre- and post-processing
#sr.setModel("edsr", 3)
sr.setModel("lapsrn",2)

# Upscale the image
result = sr.upsample(image)

# Save the image
cv2.imwrite("/content/TF-LapSRN/result/upscaled13.png", result)
```

```python
import cv2
import matplotlib.pyplot as plt
# Read image
img = cv2.imread("/content/EDSR_Tensorflow/images/input.png")
plt.imshow(img[:,:,::-1])
plt.show()
```

# **EDSR for super resolution(x4)**

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
 
path = "/content/EDSR_Tensorflow/models/EDSR_x4.pb"
 
sr.readModel(path)
 
sr.setModel("edsr",4)
 
result = sr.upsample(img)
 
# Resized image
resized = cv2.resize(img,dsize=None,fx=4,fy=4)
 
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
# Original image
plt.imshow(img[:,:,::-1])
plt.subplot(1,3,2)
# SR upscaled
plt.imshow(result[:,:,::-1])
plt.subplot(1,3,3)
# OpenCV upscaled
plt.imshow(resized[:,:,::-1])
plt.show()
```
